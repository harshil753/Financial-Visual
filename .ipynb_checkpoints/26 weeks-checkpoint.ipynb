```python
timeseries = timeseries[timeseries['Date'] > 20079999]
```

```python
pricedrop.head()
```

```python
pricedrop["Date"] = pricedrop["Date"].str.replace('-','')
pricedrop["Date"] = pricedrop["Date"].apply(pd.to_numeric, errors='coerce')
```

```python
pricedrop = pricedrop[pricedrop['Date'] > 20079999]
pricedrop
```

```python
# getting the index numbers of all the instances where the price dropped more than 10%
# this is so we can later use those index numbers to root out price drops that occur too close to each other
# in our case within 4 weeks of each other

index_values=[]
counter = 0
ticker = []

for index, row in pricedrop.iterrows():
    dropdate=row["Date"]
    droptick=row["Ticker"]
    counter = counter+1
    print(counter)
    droplist = timeseries[(timeseries["Ticker"]==droptick)]
    for index, line in droplist.iterrows():
        if (dropdate == line["Date"]):
            ticker.append(line["Ticker"])
            index_values.append(line["Unnamed: 0"])
import winsound
duration = 1000  # millisecond
freq = 440  # Hz
winsound.Beep(freq, duration)
```

```python
index_df = pd.DataFrame({
    "Ticker": ticker,
    "Index": index_values
})
```

```python
index_df
```

```python
len(index_df)
```

```python
# here we're removing all rows which have a second price drop occuring within 4 weeks of the previous price drop
# this is because we will consider that whole period one drop period

cur = 0
prev = 0
for index, row in index_df.iterrows():
    cur = row["Index"]
    if((cur - prev)<=4):
        index_df.drop(index, inplace=True)
    prev = cur
import winsound
duration = 1000  # millisecond
freq = 440  # Hz
winsound.Beep(freq, duration)
```

```python
len(index_df)
```

```python
# this is where we find out the time series records for each of the 26 weeks prior to the drop 

op = []
cl = []
da = []
ti = []
vo = []

counter = 0
for index, row in index_df.iterrows():
    counter = counter +1
    print(counter)
    item = row["Index"]
    prev = item + 26 
    droptick=row["Ticker"]
    droplist = timeseries[(timeseries["Ticker"]==droptick)]
    for index, row in droplist.iterrows():
        for x in range(item+1, prev+1):
            if(row["Unnamed: 0"] == x):
                op.append(row["Weekly Open"])
                cl.append(row["Weekly Close"])
                da.append(row["Date"])
                ti.append(row["Ticker"])
                vo.append(row["Volume"])

import winsound
duration = 1000  # millisecond
freq = 440  # Hz
winsound.Beep(freq, duration)
```

```python
prev_twenty_six = pd.DataFrame({
    "Open":op,
    "Close":cl,
    "Date":da,
    "Ticker":ti,
    "Volume":vo
})
```

```python
prev_twenty_six=prev_twenty_six[["Ticker","Date","Open","Close","Volume"]]
prev_twenty_six.head()
```

```python
prev_twenty_six.to_csv("26 weeks before 2.csv")
```
