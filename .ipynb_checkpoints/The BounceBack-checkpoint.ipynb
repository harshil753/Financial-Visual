```python
import pandas as pd
import numpy as np
import winsound
pd.set_option('display.max_rows', 1000)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
time = pd.read_csv("Pricing Time Series 2.csv")
lows = pd.read_csv("low points 2.csv")
```

```python
time = time.iloc[::-1]
```

```python
time["Date"] = time["Date"].str.replace("-","")
time["Date"] = time["Date"].apply(pd.to_numeric, errors='coerce')
```

```python
time
```

```python
lows = lows.sort_index(ascending=False)
lows
```

```python
limit = 0 
ticker = ""
close = []
date = []
stock = []
open_price = []
volume = []
num_weeks = []
counter = 0
bounceback_score = []
score = 0
low_price = []

for index, row in lows.iterrows():
    counter = counter + 1
    print(counter)
    droptick = row["Ticker"]
    dropvalue = row["Close"]
    dropdate = row["Date"]
    rebound = (dropvalue * 1.15)
    droplist = time[(time["Ticker"] == droptick)]
    num_week = 0
    for index, line in droplist.iterrows():
        if(line["Date"]>dropdate):
            num_week = num_week+1
            if(line["Weekly Close"]>=rebound):
                close.append(line["Weekly Close"])
                date.append(line["Date"])
                stock.append(line["Ticker"])
                open_price.append(line["Weekly Open"])
                volume.append(line["Volume"])
                num_weeks.append(num_week)
                low_price.append(dropvalue)
                if(num_week<=1):
                    score = 4 
                    bounceback_score.append(score)
                    break
                elif(num_week>1 & num_week<=3):
                    score = 3
                    bounceback_score.append(score)
                    break
                elif(num_week>3 & num_week<=7):
                    score = 2
                    bounceback_score.append(score)
                    break
                elif(num_week>7 & num_week<=16):
                    score = 1
                    bounceback_score.append(score)
                    break
                elif(num_week>16):
                    score = 0
                    bounceback_score.append(score)
                    break
            else:
                continue
        else:
            continue
            
winsound.Beep(440,500)
```

```python
df = pd.DataFrame({
    "Ticker":stock,
    "open":open_price,
    "close":close,
    "volume":volume,
    "date":date,
    "num of weeks":num_weeks,
    "bounceback score":bounceback_score,
    "lowest price":low_price
})
df
```
