```python
# list of amex companies
amex = pd.DataFrame(pd.read_csv("amex.csv"))
amex = amex[["Symbol","Name","MarketCap","Sector","industry"]]
amex["Symbol"] = amex["Symbol"].str.strip()
amex
```

```python
# amex monthly data api call
date=[]
monthly_open=[]
monthly_close=[]
ticker=[]
counter=0
for index, row in amex.iterrows():
    url = "https://www.alphavantage.co/query?"
    counter=counter+1
    function = "TIME_SERIES_MONTHLY"
    symbol = row["Symbol"]
    target_url = url+"function="+function+"&symbol="+symbol+"&apikey="+key
    print(symbol)
    print(counter)
    print("______________________________________________________________________")
    print(target_url)
    limit=0
    try:
        info = req.get(target_url).json()
        while limit <= 60:
            for item in info["Monthly Time Series"]: 
                    ticker.append(symbol)
                    date.append(item)
                    monthly_open.append(info["Monthly Time Series"][item]["1. open"])
                    monthly_close.append(info["Monthly Time Series"][item]["4. close"])
                    limit = limit+ 1
    except KeyError:
        pass
    except KeyboardInterrupt:
        raise
```

```python
# finding the earnings per share

actual_eps = [] 
estimated_eps = []
fiscal_period = []
stock_ticker = []
counter = 0
for index, row in amex.iterrows():
    counter = counter + 1
    symbol = row["Symbol"]
    url = "https://api.iextrading.com/1.0/stock/"+symbol+"/earnings"
    print(counter)
    print(symbol)
    print(url)
    try:
        info = req.get(url).json()
        for x in range(0,4):
            stock_ticker.append(symbol)
            actual_eps.append(info["earnings"][x]["actualEPS"])
            estimated_eps.append(info["earnings"][x]["estimatedEPS"])
            fiscal_period.append(info["earnings"][x]["fiscalPeriod"])
    except:
        pass
```

```python
# finding other useful metrics

pe_ratio = []
market_cap = []
ytdchange = []
stocksymbol = []
counter = 0 

for index, row in amex.iterrows():
    counter = counter + 1
    symbol = row["Symbol"]
    url = "https://api.iextrading.com/1.0/stock/"+symbol+"/quote"
    print(counter)
    print(symbol)
    print(url)
    try:
        info = req.get(url).json()
        stocksymbol.append(symbol)
        pe_ratio.append(info["peRatio"])
        market_cap.append(info["marketCap"])
        ytdchange.append(info["ytdChange"])
    except:
        pass
```

```python
metrics = pd.DataFrame({
    "Ticker": stocksymbol,
    "PE Ratio":pe_ratio,
    "Market Cap":market_cap,
    "Year To Date Change": ytdchange
})
metrics
```
